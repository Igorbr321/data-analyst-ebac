## 1\. Preços `outliers` de diamante

```python
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```python
# gráfico do exercício 1

diamonds = sns.boxplot(x=data['price'])
diamonds.set(title='Preços Diamantes', xlabel='Preço');
```

**Insight do gráfico 1**: A maioria dos preços dos diamantes está concentrada em uma faixa mais baixa, enquanto há um pequeno número de diamantes que são muito mais caros, considerados outliers.

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```python
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```python
# gráfico do exercício 2

diamonds = data[['price', 'carat', 'clarity']]

with sns.axes_style('whitegrid'):

    grafico = sns.scatterplot(data=diamonds, x='price', y='carat', hue='clarity', palette='pastel')
    grafico.set(title='Correlação do Preço por Peso', xlabel='Preço', ylabel='Peso');
    grafico.get_legend().set_title('Clareza');
```

**Insight do gráfico 2**: O gráfico mostra que o peso é um dos principais determinantes do preço de um diamante, mas outros fatores, como a clareza, também desempenham um papel importante.

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```python
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```python
# gráfico do exercício 3

cars = data['weight']
grafico = sns.histplot(data=cars, kde=True)
grafico.set(title='Distribuição Contínua Aproximada do Peso de Carros', xlabel='Peso', ylabel='Frequência');
```

**Insight do gráfico 3**: A distribuição apresenta uma leve assimetria para a direita, indicando a presença de alguns carros mais pesados, possivelmente SUVs ou caminhões, que são menos comuns. Essa distribuição sugere que, durante o período analisado, os carros de peso médio eram predominantes, com uma menor diversidade de veículos mais pesados.
